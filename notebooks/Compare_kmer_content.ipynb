```python
infile_sequins_only = '../../run/sequins_transcripts.tsv'
infile_sequins_gencode = '../../run/gencode_sequins.tsv'
infile_sequins_basic_gencode = '../../run/gencode_basic_sequins.tsv'
infile_sirvs_only = '/home/FILESERVER5/proteomics/tss38/Downloads/SIRV_transcriptome.tsv'
```

```python
import pandas as pd
%load_ext rpy2.ipython
```

```python
sequins_only_df = pd.read_table(infile_sequins_only, sep='\t', index_col=0)
sequins_only_df['transcriptome'] = 'sequins_only'
sequins_only_df['type'] = 'sequin'
print(sequins_only_df.head())

sequins_gencode_df = pd.read_table(infile_sequins_gencode, sep='\t', index_col=0)
sequins_gencode_df['transcriptome'] = 'plus_gencode'
sequins_gencode_df['type'] = ['gencode' if 'ENST' in x else 'sequin' for x in sequins_gencode_df.index]
print(sequins_gencode_df.head())

sequins_basic_gencode_df = pd.read_table(infile_sequins_basic_gencode, sep='\t', index_col=0)
sequins_basic_gencode_df['transcriptome'] = 'plus_gencode_basic'
sequins_basic_gencode_df['type'] = ['gencode' if 'ENST' in x else 'sequin' for x in sequins_basic_gencode_df.index]
print(sequins_basic_gencode_df.head())


sirvs_only_df = pd.read_table(infile_sirvs_only, sep='\t', index_col=0)
sirvs_only_df['transcriptome'] = 'SIRV_only'
sirvs_only_df['type'] = 'SIRV'
print(sirvs_only_df.head())

#final_df = pd.concat([sequins_only_df, sequins_gencode_df])
final_df = pd.concat([sequins_only_df, sequins_gencode_df, sequins_basic_gencode_df, sirvs_only_df])
```
